The aim is to predict the activity of person whether s/he is STANDING, SITTING, LAYING, WALKING, WALKING_DOWNSTAIRS, WALKING_UPSTAIRS based on 563 parameters.

This data set condist of 3609 rows, 563 columns and 2,031,867 observations.

```python
Activity = {'LAYING': 1,'STANDING': 2,'SITTING': 3,'WALKING': 4,'WALKING_UPSTAIRS': 5,'WALKING_DOWNSTAIRS': 6}
```

```python
data.activity = [Activity[item] for item in data.activity]
```

```python
data.head()
```

```python
X = pd.DataFrame()
def generate():
    X = data.loc[:,data.columns!='activity']
    return X
X = generate()
```

```python
y = pd.DataFrame()
def generate():
    y = data.loc[:,data.columns=='activity']
    return y
y = generate()
```

```python
from sklearn.preprocessing import StandardScaler
def generate():
    sc = StandardScaler()
    sc.fit(X)
    return sc.transform(X)
X = generate()
```

```python
from sklearn.model_selection import train_test_split
def generate():
    return train_test_split(X, y, test_size=0.25, random_state=0)
X_train, X_test, y_train, y_test = generate()
```

```python
def generate():
    print (X_train.shape)
    print (y_train.shape)
    return
generate()
```

```python
def generate():
    print (X_test.shape)
    print (y_test.shape)
    return
generate()
```

```python
from sklearn.svm import SVC
from sklearn import metrics
def generate():
    svc = SVC(random_state=0)
    svc.fit(X_train,y_train)
    y_pred = svc.predict(X_test)
    accuracy_score = metrics.accuracy_score(y_test,y_pred)
    print ('accuracy score is: -',accuracy_score)
generate()
```
